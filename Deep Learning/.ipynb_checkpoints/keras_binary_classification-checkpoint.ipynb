```python
import pandas as pd
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df
```

```python
df.info()
```

```python
df.columns
```

```python
y = df['Exited']
```

```python
X = df[['RowNumber', 'CustomerId', 'Surname', 'CreditScore', 'Geography',
       'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary']]
```

```python
X
```

## Feature Engineering: 
Data Preprocessing, Data Imputation, One Hot Encoding, Feature Elimination, etc 

```python
X.drop(['RowNumber'], axis=1, inplace=True)
```

```python
X.drop(['CustomerId'], axis=1, inplace=True)
```

```python
X.drop(['Surname'], axis=1, inplace=True)
```

```python
X
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X
```

```python
# test and train split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, shuffle=True)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential()
```

```python
model.get_config()
```

```python
model.add(tf.keras.layers.Dense(
    units=64,
    input_dim=11,
    activation='relu',
))
```

```python
model.add(tf.keras.layers.Dense(
    units=32,
    activation='relu',
    kernel_regularizer=tf.keras.regularizers.l2(l=0.01)
))
```

```python
#model.get_layer("dense_1").get_weights()
```

```python
model.add(tf.keras.layers.Dense(
    units=1,
    activation='sigmoid',
))
```

```python
model.summary()
```

```python
my_learning_rate = 0.01
my_epochs = 100
```

```python
model.compile(
    optimizer= tf.keras.optimizers.Adam(learning_rate = my_learning_rate), 
    loss='binary_crossentropy', 
     metrics= ['accuracy'])
```

```python
#model.get_weights()
```

```python
history=model.fit(x=X_train, y=y_train, epochs=my_epochs,validation_split=0.33, shuffle=True)
```

```python
epochs = history.epoch
```

```python
hist = pd.DataFrame(history.history)
```

```python
hist
```

```python
loss = hist['loss']
vloss = hist['val_loss']
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_the_loss_curve(epochs, loss, vloss):
    plt.figure()
    plt.xlabel("Epoch")
    plt.ylabel('Loss')
    plt.plot(epochs, loss, label='Loss')
    plt.plot(epochs, vloss, label='Val Loss')
    plt.legend()
    plt.ylim([loss.min(), loss.max()])
    plt.show
```

```python
plot_the_loss_curve(epochs, loss, vloss)
```
