```python
# Bad Algorithm: Time Complexity is O(n)
def lw(j):
    x = 0
    for i in range(j):
        x = i + x
    print(x)
```

```python
# Good Alorithm: Time Complexity = O(1)
def lw2(j):
    x = 0
    x = (j * (j+1))/2
    print(x)
```

```python
lw(100)
```
