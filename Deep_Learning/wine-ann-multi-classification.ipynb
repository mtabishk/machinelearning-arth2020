```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/Wine.csv', header=None)
```

```python
df.columns = [  'class'
                 ,'alcohol'
                ,'malicAcid'
                ,'ash'
                ,'ashalcalinity'
                ,'magnesium'
                ,'totalPhenols'
                ,'flavanoids'
                ,'nonFlavanoidPhenols'
                ,'proanthocyanins'
                ,'colorIntensity'
                ,'hue'
                ,'od280_od315'
                ,'proline'
                ]
```

```python
df
```

```python
df.columns
```

```python
y = df['class']
```

```python
y.value_counts()
```

```python
y_cat = pd.get_dummies(y)
```

```python
X = df.drop('class', axis=1)
```

```python
X
```

```python
X.info()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x='alcohol', y = 'class', data=df)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.get_config()
```

```python
X.shape
```

```python
model.add(Dense(
    units=32, 
    input_shape=(13,), 
    activation='relu', 
    kernel_initializer='he_normal'
))
```

```python
model.add(Dense(
    units=16, 
    activation='relu', 
    kernel_initializer='he_normal'
))
```

```python
model.add(Dense(
    units=8, 
    activation='relu', 
    kernel_initializer='he_normal'
))
```

```python
model.add(Dense(
    units=3, 
    activation='softmax', 
    kernel_initializer='he_normal'
))
```

```python
model.summary()
```

```python
from tensorflow.keras.losses import categorical_crossentropy
from tensorflow.keras.optimizers import RMSprop
```

```python
model.compile(
    optimizer=RMSprop(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X, y_cat, epochs=200)
```

```python
model.history.history
```

```python
loss = model.history.history['loss']
```

```python
loss_pd = pd.DataFrame(loss)
```

```python
loss_pd.plot()
```

```python
accuracy =  model.history.history['accuracy']
```

```python
accuracy_pd = pd.DataFrame(accuracy)
```

```python
accuracy_pd.plot()
```
